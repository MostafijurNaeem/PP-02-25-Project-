```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Initialize variables
expression = ""

# Function to handle button clicks
def on_click(change, value):
    global expression
    if value == "=":
        try:
            # Evaluate the expression and display the result
            expression = str(eval(expression))
        except:
            expression = "Error"
    elif value == "C":
        # Clear the expression
        expression = ""
    else:
        # Append the clicked button's value to the expression
        expression += value
    # Update the display
    display_output.value = expression

# Create display output area
display_output = widgets.Text(value="", description="", layout=widgets.Layout(width="250px", height="50px"))
display_output.style = {"font_size": "20px", "font_weight": "bold"}

# Button colors
button_colors = {
    "numbers": "black",
    "operators": "blue",
    "clear": "red",
    "equals": "green",
}

# Create buttons
buttons = []
button_labels = [
    ["7", "8", "9", "/"],
    ["4", "5", "6", "*"],
    ["1", "2", "3", "-"],
    ["C", "0", "=", "+"],
]

for row in button_labels:
    button_row = []
    for label in row:
        color = (
            button_colors["numbers"] if label.isdigit()
            else button_colors["operators"] if label in "+-*/"
            else button_colors["clear"] if label == "C"
            else button_colors["equals"]
        )
        button = widgets.Button(
            description=label,
            layout=widgets.Layout(width="60px", height="60px"),
            style={"button_color": color},
        )
        button.on_click(lambda change, value=label: on_click(change, value))
        button_row.append(button)
    buttons.append(button_row)

# Create a grid for buttons
grid = widgets.GridBox(
    children=[btn for row in buttons for btn in row],
    layout=widgets.Layout(grid_template_columns="repeat(4, 60px)", grid_gap="5px"),
)

# Display the calculator
display(display_output)
display(grid)
```
